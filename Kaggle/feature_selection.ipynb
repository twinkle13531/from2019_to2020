```python
#データを読み込む
import pandas as pd
data=pd.read_csv('breast_cancer.csv')
data.head()
```

```python
#columnsの詳細説明。どのくらい揃っているのか、データ型などが記してある
data.info()
```

```python
col = data.columns       # .columns gives columns names in data 
print(col)
```

```python
y = data.diagnosis                          # M or B 
list = ['Unnamed: 32','id','diagnosis'] #data.info()で見たcolumnsのうち絶対いらないのを除く
x = data.drop(list,axis = 1 ) #listのcolumn(列指定)に対して全データを削除する(axis=1方向⇔列方向) #列の削減
x.head()
```

```python
drop_list1 = ['perimeter_mean','radius_mean','compactness_mean','concave points_mean',
              'radius_se','perimeter_se','radius_worst','perimeter_worst','compactness_worst',
              'concave points_worst','compactness_se','concave points_se','texture_worst','area_worst']
x_1 = x.drop(drop_list1,axis = 1 )  
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score

# split data train 70 % and test 30 %
x_train, x_test, y_train, y_test = train_test_split(x_1, y, test_size=0.3, random_state=42)
```

```python
#反復特徴量選択
from sklearn.feature_selection import RFE #Recursive Feature Elimination; 再帰的特徴量削減

# Create the RFE object and rank each pixel
clf_rf_3 = RandomForestClassifier()      
rfe = RFE(estimator=clf_rf_3, n_features_to_select=5, step=1)
rfe = rfe.fit(x_train, y_train)
```

```python
print('Chosen best 5 feature by rfe:',x_train.columns[rfe.support_]) 
```

```python
#反復特徴量選択
from sklearn.feature_selection import RFECV #RFE+交差検証(CV)

# ランダムフォレスト
clf_rf_4 = RandomForestClassifier() 

#特徴量削減
rfecv = RFECV(estimator=clf_rf_4, step=1, cv=5,scoring='accuracy') #5-fold cross-validation(5分割交差検証）

#学習
rfecv = rfecv.fit(x_train, y_train)

print('Optimal number of features :', rfecv.n_features_)
print('Best features :', x_train.columns[rfecv.support_])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score of number of selected features")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()
```

```python
#モデルベース特徴量選択
from sklearn.feature_selection import SelectFromModel

# ランダムフォレスト
clf_rf_5 = RandomForestClassifier(random_state=0)
clf_rf_5.fit(x_train, y_train)

# 重要度が平均値以上の特徴量を抽出
sfm = SelectFromModel(clf_rf_5, threshold='mean', prefit=True)
transformed_x = sfm.transform(x_train)

# 30次元を低次元にする
print('Transformed X shape: {}'.format(transformed_x.shape))
```

```python
#単純変量統計
from sklearn.datasets import load_boston
from sklearn.feature_selection import SelectKBest, f_regression

boston = load_boston()
X = boston.data
y = boston.target

# 上位5つの特徴量を選択
selector = SelectKBest(score_func=f_regression, k=5) #f_regressionを指定⇔ 回帰
selector.fit(X, y)
mask = selector.get_support()  #各特徴量を選択したか否かのmaskを取得
print(boston.feature_names)
print(mask)

# 選択した特徴量の列のみ取得
X_selected = selector.transform(X)
print("X.shape={}, X_selected.shape={}".format(X.shape, X_selected.shape))
```

```python
#単純変量統計
from sklearn.datasets import load_boston
from sklearn.feature_selection import SelectPercentile, f_regression #価格と特徴量の関数を知りたいので、回帰問題です

boston = load_boston()
X = boston.data
y = boston.target #価格

selector = SelectPercentile(score_func=f_regression, percentile=40) #regression回帰、全特徴量のうち40%を選択すると宣言 
selector.fit(X, y)
mask = selector.get_support() #True, False
print(boston.feature_names) #dataのうちfeature_names
print(mask)

X_selected = selector.transform(X)
print("X.shape={}, X_selected.shape={}".format(X.shape, X_selected.shape))
```

```python
#モデルベース特徴量選択
from sklearn.datasets import load_boston
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestRegressor

boston = load_boston()
X = boston.data
y = boston.target

# estimator として RandomForestRegressor を使用。重要度が median 以上のものを選択
selector = SelectFromModel(RandomForestRegressor(n_estimators=100, random_state=42), threshold="median")    
selector.fit(X, y)
mask = selector.get_support()
print(boston.feature_names)
print(mask)

# 選択した特徴量の列のみ取得
X_selected = selector.transform(X)
print("X.shape={}, X_selected.shape={}".format(X.shape, X_selected.shape))
```

```python
#反復特徴量選択
from sklearn.datasets import load_boston
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor

boston = load_boston()
X = boston.data
y = boston.target

# estimator として RandomForestRegressor を使用。特徴量を5個選択
selector = RFE(RandomForestRegressor(n_estimators=100, random_state=42), n_features_to_select=5)
selector.fit(X, y)
mask = selector.get_support()
print(boston.feature_names)
print(mask)

# 選択した特徴量の列のみ取得
X_selected = selector.transform(X)
print("X.shape={}, X_selected.shape={}".format(X.shape, X_selected.shape))
```
